```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r = np.linspace(-1, 1, 1000)

def f(r):
  return -r

filled_marker_style = dict(marker='o', linestyle=':', markersize=9,
                           color='None',
                           markerfacecolor='white',
                           markerfacecoloralt='k',
                           markeredgecolor='k')

plt.figure(figsize=(6, 4.3))
plt.plot(r, f(r), lw=1, ls='solid', c='k', label='stable')
# plt.plot(r2, xeq(r2), lw=1, ls='dashed', c='k', label='instable')
# plt.plot(r1, np.zeros(1000), lw=1, ls='dashed', c='k')
# plt.plot(r2, np.zeros(1000), lw=1, ls='solid', c='k')
# plt.plot([0], [0], fillstyle='left', **filled_marker_style, label='bifurcation point', zorder=9)

plt.xlim(-1, 1)
plt.xlabel('quantity, $Q$', size=14)
plt.ylabel('price, $P$', size=14)
plt.title('Linear regulation of consumer demand\n', size=14)
plt.legend(frameon=False)
plt.tight_layout()
# plt.xticks([])
# plt.yticks([])
plt.savefig('linear-regulation.png', dpi=150)
```

```python
def f(r, r_max=1, k=20):
    return r_max / (1 + np.exp(-k * r))

r = np.linspace(-1, 1, 1000)

filled_marker_style = dict(marker='o', linestyle=':', markersize=9,
                           color='None',
                           markerfacecolor='white',
                           markerfacecoloralt='k',
                           markeredgecolor='k')

plt.figure(figsize=(6, 4.3))
plt.plot((r[::-1]*-1 + 1) / 2, f(r), lw=1, ls='solid', c='k', label='stable')

plt.xlim(1, 0)
plt.gca().xaxis.set_inverted(True)
plt.xlabel('density of forest, $P$', size=14)
plt.ylabel('fraction of burned trees $F$', size=14)
plt.title('Non-linear regulation of a forest fire\n', size=14)
plt.tight_layout()
plt.legend(frameon=False)
plt.savefig('logistic-regulation.png', dpi=150)
plt.show()
```

```python
# dx/dt = h + r*x - x**3
# xeq = -r*x + r**3

x1 = np.linspace(-10, 10, 1000)
x2 = np.linspace(-10, -0.6, 1000)
x3 = np.linspace(0.6, 10, 1000)
r = 1

def h(r, x):
  return -r*x + x**3

filled_marker_style = dict(marker='o', linestyle=':', markersize=9,
                           color='None',
                           markerfacecolor='white',
                           markerfacecoloralt='k',
                           markeredgecolor='k')

plt.figure(figsize=(6, 4.3))
plt.plot(h(r, x2), x2, lw=1, ls='solid', c='k', label='stable')
plt.plot(h(r, x3), x3, lw=1, ls='solid', c='k')
plt.plot(h(r, x1), x1, lw=1, ls='dashed', c='k', label='instable')
# plt.plot([-0.383], [0.57], fillstyle='left', **filled_marker_style, label='bifurcation\npoint', zorder=9)
# plt.plot([0.383], [-0.57], fillstyle='left', **filled_marker_style, zorder=9)
plt.xlim(-1, 1)
plt.ylim(-1.5, 1.5)
plt.title('Hysteretic regulation of a shallow lake\n', size=14)
plt.xlabel('waste loading, $W$', size=14)
plt.ylabel('turbidity, $T$', size=14)
plt.legend(frameon=False)
plt.tight_layout()
plt.savefig('hysteretic-regulation.png', dpi=150)
```

```python
# dx/dt = r*x - x**3
# xeq = r**2

r1 = np.linspace(0.0001, 1, 1000)
r2 = np.linspace(-1, 0, 1000)

def xeq1(r):
  return r**(1/2)

def xeq2(r):
  return -r**(1/2)

filled_marker_style = dict(marker='o', linestyle=':', markersize=9,
                           color='None',
                           markerfacecolor='white',
                           markerfacecoloralt='k',
                           markeredgecolor='k')

plt.figure(figsize=(6, 4.3))
plt.plot(r1*-1, xeq1(r1), lw=1, ls='solid', c='k', label='stable')
plt.plot(r1*-1, xeq2(r1), lw=1, ls='solid', c='k')
plt.plot(r1*-1, np.zeros(1000), lw=1, ls='dashed', c='k', label='instable')
plt.plot(r2*-1, np.zeros(1000), lw=1, ls='solid', c='k')
# plt.plot([0], [0], fillstyle='left', **filled_marker_style, label='bifurcation point', zorder=9)

plt.xlim(-1, 1)
plt.gca().xaxis.set_inverted(True)
plt.title('Probabilistic regulation of a magnet\n', size=14)
plt.xlabel('Temperature, $t$', size=14)
plt.ylabel('Magnetization, $M$', size=14)
plt.legend(frameon=False)
plt.tight_layout()
plt.savefig('probabilistic-regulation.png', dpi=200)
```

```python
def logistic(x, r):
  return r * x * (1 - x)

n = 10000
x = 1e-5 * np.ones(n)
r = np.linspace(2.2, 3.7, n)

iterations = 1000
last = 900

plt.figure(figsize=(6, 4.3))

for i in range(iterations):
  x = logistic(x, r)
  if i >= (iterations - last):
    plt.plot(r, x, ',k', lw=1, alpha=0.01)

plt.xlim(2.2, 3.7)
plt.title('Chaotic regulation of bacterial growth\n', size=14)
plt.xlabel('Population density, $P$', size=14)
plt.ylabel('Growth rate, $R$', size=14)
plt.tight_layout()
plt.savefig('chaotic-regulation.png', dpi=150)
```
